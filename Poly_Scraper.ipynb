```python
from lxml import html
from json import dump,loads
from requests import get
import json
from re import sub
from dateutil import parser as dateparser
from time import sleep
import pandas as pd
```

```python
listofprof = "http://polyratings.com/list.php"
headers = {'User-Agent': 'Mozilla/5.0 (Windows; U; Windows NT 6.1; en-US) AppleWebKit/534.13 (KHTML, like Gecko) Chrome/9.0.597.0 Safari/534.13'}
response = get(listofprof, headers = headers, verify=False, timeout=30)
cleaned_response = response.text.replace('\x00', '')
parser = html.fromstring(cleaned_response)

```

```python
XPATH_PROF = './/a[@class="no-link-highlight text-muted filterable"]'

XPATH_AGG = './/span[@class="glyphicon glyphicon-star star-color"]'
```

```python
urls = []
professors = parser.xpath(XPATH_PROF)
counter = 1
for i in professors:
    urls.append([i.attrib['href'],parser.xpath('/html/body/div/a[{0}]/div/div/div[2]/h4/text()'.format(counter))[0]
                 .replace(" ","").replace("\n","")])
    counter += 1
```

```python
data = []
for url in urls:
    response = get(url[0], headers = headers, verify=False, timeout=30)
    cleaned_response = response.text.replace('\x00', '')
    parser = html.fromstring(cleaned_response)
    XPATH_NAME = './/h1[@class="text-primary"]//text()'
    try:
        prof_name = parser.xpath(XPATH_NAME)[0]
    except:
        #page does not exist
        continue
    XPATH_CLASSES = './/div[@class="col-xs-12 col-sm-8"]'
    agg_rating = parser.xpath('/html/body/div[1]/div[4]/span/h2')[0].text_content()
    list_classs = parser.xpath(XPATH_CLASSES)[0].xpath("//section")
    review_list = []
    for i in list_classs:
        class_name = i.attrib['id']
        counter = 1
        res = ['dummy']
        while res != []:
            for counter1 in range(1,3):
                XPATH_STUDS = ('//*[@id="{0}"]/div[{1}]/div/div[{2}]//text()'.format(class_name,
                                                             counter,
                                                             counter1))
                if counter1 == 1:
                    #print(XPATH_STUDS)
                    res = parser.xpath(XPATH_STUDS)
                    if res == []:
                        continue
                    clean_attribute = [i.replace("\n","") for i in res]
                    standing = clean_attribute[0]
                    grade = clean_attribute[1]
                    reason = clean_attribute[2]
                    date = clean_attribute[3]
                if counter1 == 2:
                    #print(XPATH_STUDS)
                    res = parser.xpath(XPATH_STUDS)
                    if res == []:
                        continue
                    review = res[0].replace("\n","")
            if res != []:
                review_dict = {
                                'class' : class_name,
                                'standing' : standing,
                                'grade' : grade,
                                'reason' : reason,
                                'date' : date,
                                'review' : review                         
                }
                review_list.append(review_dict)
            counter += 1
    data.append({
                        "prof_name": prof_name,
                        "agg_rating" : agg_rating,
                        "subject" : url[1],
                        "reviews" : review_list
    })
    print(prof_name)
                
```

```python
f = open("polyrating.json","w")
dump(data, f, indent=4)
f.close
```
